```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyarrow as pa
import pyarrow.parquet as pq
import geopandas as gpd
import seaborn as sns

#Packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
from scipy import stats
pd.set_option('display.max_columns', None)

from scipy.stats import skew, kurtosis

import pygris
import shapely
```

```python
states = pygris.states()
states_list = sorted(states['STUSPS'].unique())
```

```python
BG_list_2020 = []

for i in states_list:
    BG_list_2020.append(pygris.block_groups(state = i, year = 2020))
```

```python
BG_list_2010 = []

for i in states_list:
    BG_list_2010.append(pygris.block_groups(state = i, year = 2010))
```

```python
BG_list_2000 = []

for i in states_list:
    BG_list_2000.append(pygris.block_groups(state = i, year = 2000))
```

```python
BG_list_1990 = []

for i in states_list:
    BG_list_1990.append(pygris.block_groups(state = i, year = 1990, cb = True))
```

```python
temp2020 = pd.concat(BG_list_2020, ignore_index=True)
temp2010 = pd.concat(BG_list_2010, ignore_index=True)
temp2000 = pd.concat(BG_list_2000, ignore_index=True)
temp1990 = pd.concat(BG_list_1990, ignore_index=True)
```

```python
temp2020 = temp2020[['GEOID', 'geometry']]
temp2020['year'] = 2020

temp2010 = temp2010.rename(columns={"GEOID10": "GEOID"})

temp2010 = temp2010[['GEOID', 'geometry']]
temp2010['year'] = 2010

temp2000 = temp2000[['GEOID', 'geometry']]
temp2000['year'] = 2000

# temp2000 = temp2000.rename(columns={"BKGPIDFP00": "GEOID"})

temp1990 = temp1990[['GEOID', 'geometry']]
temp1990['year'] = 1990
```

```python
# Find the centroid for each geometry
centroids = visualize.geometry.centroid

# Filter the rows where the y-coordinate (latitude) of the centroid is greater than 42
visualize = visualize[(centroids.y < 55) & (centroids.y > 22) & (centroids.x < -50) & (centroids.x > -140)]

visualize = visualize[visualize['GEOID']!='020160001001']

# Convert the DataFrame to a GeoDataFrame
gdf = gpd.GeoDataFrame(visualize, geometry='geometry')

# Set the current CRS to WGS 84 (assuming your data is in latitude and longitude)
gdf.crs = "EPSG:4326"

# Project the GeoDataFrame to a suitable CRS (using Web Mercator here for demonstration)
gdf_projected = gdf.to_crs(epsg=3857)

# Compute the average area in square meters
average_area_m2 = gdf_projected.geometry.area.sum()
# Convert the area from square meters to square kilometers
average_area_km2 = average_area_m2 / 1e6
print(f"Total Area: {average_area_km2} square kilometers")

# Plot using the 'viridis' colormap
gdf.plot(edgecolor="black", cmap='Reds', figsize=(15, 15))

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Using Census BlockGroup geometries shapefile for 2020')

plt.show()
```
